```python
chosun_eilbo_cos_sim_tfidf_ = pd.read_csv(r'C:\bigdata_git\bigdata_thunder\bigdata2\chosun_tail_100_cos_sim_tfidf.csv', encoding='utf-8')
korean_economy_cos_sim_tfidf_ = pd.read_csv(r'C:\bigdata_git\bigdata_thunder\bigdata2\korean_tail_100_cos_sim_tfidf.csv', encoding='utf-8')
mail_economy_cos_sim_tfidf_ = pd.read_csv(r'C:\bigdata_git\bigdata_thunder\bigdata2\mail_tail_100_cos_sim_tfidf.csv', encoding='utf-8')
midlle_eilbo_cos_sim_tfidf_ = pd.read_csv(r'C:\bigdata_git\bigdata_thunder\bigdata2\middle_tail_100_cos_sim_tfidf.csv', encoding='utf-8')
money_today_cos_sim_tfidf_ = pd.read_csv(r'C:\bigdata_git\bigdata_thunder\bigdata2\money_tail_100_cos_sim_tfidf.csv', encoding='utf-8')
```

```python
new_chosun_cos_sim_array = []
for i in chosun_eilbo_cos_sim_tfidf_.values:
    result = []
    for j in range(len(i)):
        if j != 0:
            result.append(i[j])
    new_chosun_cos_sim_array.append(result)
print(len(new_chosun_cos_sim_array))
new_chosun_cos_sim_list = []
for array in range(len(new_chosun_cos_sim_array)):
    for i in range(len(new_chosun_cos_sim_array[array])):
        if i == array:
            new_chosun_cos_sim_list.append(new_chosun_cos_sim_array[array][i])
print(new_chosun_cos_sim_list)
```
